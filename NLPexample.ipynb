```python
import nltk
import numpy as np
import random
import string # to process standard python strings
import nltk.corpus
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')
from nltk.corpus import stopwords
stopwords.words('spanish')
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
f=open('BSAS.txt','r',errors = 'ignore')
raw=f.read()
raw=raw.lower()# minuscula
sent_tokens = nltk.sent_tokenize(raw)# convertir en lista de oraciones 
word_tokens = nltk.word_tokenize(raw)# convertir en lista de palabras
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
sent_tokens = lemmatizer.lemmatize(str(word_tokens))
sent_tokens[:2]
```

```python
word_tokens = [''.join(c for c in s if c not in string.punctuation) for s in word_tokens]
word_tokens = [s for s in word_tokens if s]
```

```python
clean_tokens = word_tokens[:]

sr = stopwords.words('spanish')

for token in word_tokens:

    if token in stopwords.words('spanish'):

        clean_tokens.remove(token)
```

```python
clean_tokens
```

```python
vectorizer = TfidfVectorizer()
analyze = vectorizer.build_analyzer()
print(analyze(str(clean_tokens)))
```

```python
X = vectorizer.fit_transform(word_tokens)
print(vectorizer.get_feature_names())
```

```python
freq = nltk.FreqDist(clean_tokens)

for key,val in freq.items():

    print (str(key) + ':' + str(val))
```

```python
import heapq
most_freq = heapq.nlargest(200, freq, key=freq.get)
```
